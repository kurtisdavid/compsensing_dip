```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 5)
for i in range(5):
    me = np.load('noisy_'+str(i)+'.npy')
    ax[i].imshow(me.reshape(7,7))
```

```python
fig, ax = plt.subplots(1, 5)
for i in range(5):
    me = np.load('im'+str(i+1)+'.npy')
    ax[i].imshow(me.squeeze())
```
